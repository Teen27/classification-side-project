```python
# since only a few records have null values in training set, replace null values with mean and mode for numerical and categorical
# variables respectively
def preprocess(df):
    import pandas as pd
    # split data into numeric and categorical parts
    df_cols = df.columns
    num_df = df._get_numeric_data()
    cat_df = df[list(set(df_cols)- set(num_df.columns))]
    
    # convert categorical vaues to string type
    for i in cat_df.columns:
        cat_df[i] = cat_df[i].astype('str')
    
    # replace null values with mean and mode for numerical and categorical variables respectively
    n_df = num_df.fillna(num_df.mean())
    c_df = cat_df.fillna(cat_df.mode())
    
    # clean categorical variables, replace nan values with mead/modes since there are few of them
    c_df['x41'] = c_df['x41'].str.replace('$','').astype('float')
    c_df['x41'].fillna(c_df['x41'].mean(), inplace = True)
    
    c_df['x45'] = c_df['x45'].str.replace('nan',str(c_df['x45'].mode()[0])).str.replace('-0.0%','0.0%').str.replace('%','').astype('float')
    c_df['x45'] /= 100
    #print('1/2')
    c_df['x34'] = c_df['x34'].str.replace('nan', c_df['x34'].mode()[0])
    c_df['x68'] = c_df['x68'].str.replace('nan', c_df['x68'].mode()[0])
    c_df['x93'] = c_df['x93'].str.replace('nan', c_df['x93'].mode()[0])
    #print('2/3')
    vals_to_replace = {'fri':'friday', 'thur':'thurday', 'wed':'wednesday'}
    c_df['x35'] = c_df['x35'].fillna(c_df['x35'].mode()[0]).replace(vals_to_replace)
    c_df['x35'] = c_df['x35'].str.replace('nan', c_df['x35'].mode()[0])
    
    # for nominal variables, do one-hot encoder; for ordinal variables, convert it to integer
    weekdays_to_int = {'monday': 1, 'tuesday': 2, 'wednesday':3, 'thurday':4, 'friday':5}
    c_df['x35'] = c_df['x35'].replace(weekdays_to_int)
    
    months_to_int = {'January':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'Jun':6,'July':7,'Aug':8,'sept.':9,'Oct':10,'Nov':11,'Dev':12}
    c_df['x68'] = c_df['x68'].replace(months_to_int)
    
    clean_df = pd.concat([n_df, c_df], axis = 1)
    clean_df = pd.get_dummies(clean_df)
    return(clean_df, n_df, c_df)
```

```python
clean_train, num_train, cat_train = preprocess(train)
```

```python
clean_test, num_test, cat_test = preprocess(test)
```

```python
# check if we still have null values in the columns
print(clean_train.columns[clean_train.isna().any()].tolist())
print(clean_test.columns[clean_test.isna().any()].tolist())
```

```python
X_train, X_test, y_train, y_test = train_test_split(clean_train.iloc[:,clean_train.columns != 'y'],
                                                    clean_train['y'],random_state = 0, test_size = 0.2)


random_grid = {'n_estimators': [100, 500, 1000, 1500, 2000],
               'max_features': [None, 'sqrt'],
               'max_depth': [10, 50, 100]
              }

rf = RandomForestClassifier()
grid_rf = GridSearchCV(rf, param_grid=random_grid, scoring='roc_auc', cv=5, n_jobs=-1, verbose= 2) # given we will use AUC
grid_rf.fit(X_train, y_train)
grid_rf.best_params_
```

```python
# Don't forget we have class imbalance problem, we can adjust that by changing the class weight to be 'balanced'
# Due to the limited computer setting and time, I was only able to grid search a few values
# Further increase the n_estimators to improve performance
X_train, X_test, y_train, y_test = train_test_split(clean_train.iloc[:,clean_train.columns != 'y'],
                                                    clean_train['y'],random_state = 0, test_size = 0.2)

best_rf = RandomForestClassifier(max_depth= 50, max_features= 'sqrt', n_estimators= 1500, class_weight='balanced')
best_rf.fit(X_train, y_train)
pred_values = best_rf.predict_proba(X_test)
roc_auc_score(y_test, pred_values[:,1])
# AUC is 0.9831650005915059 in testing set
```

```python
# predicting in test set
pred_test_values = best_rf.predict_proba(clean_test)
result1 = pred_test_values[:,1]
# output random forest prediction result to 'result1.csv'
np.savetxt('result1.csv', result1)
```
